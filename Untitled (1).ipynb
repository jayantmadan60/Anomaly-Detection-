```python
# open the data file and reading ot in to the JSON format with 'utf-8' encoding
file = open('transactions.json','r',encoding='utf-8')
transactions = []
for line in file.readlines():
    dic = json.loads(line)
    transactions.append(dic)
    
# print total number of transactions in a dataset
print (len(transactions))
```

```python
# converting the JSON dictonary in to the dataframe
df_transactions = pd.DataFrame(transactions)
```

```python
# checking number of columns in dataframe
df_transactions.columns
```

```python
# checking the column types and info in the dataframe
for i in df_transactions.columns:
    print ('%s = %s' %(i,(df_transactions[i].dtype)))
print ('\n #########################################')
df_transactions.describe()
```

```python
# changing column types in the transactions dataframe       ("posConditionCode": int , "posEntryMode": int)
df_transactions = df_transactions.astype({"accountNumber": int, 
                                          "customerId": int,
                                          "cardCVV": int,
                                          "enteredCVV": int,
                                          "cardLast4Digits": int})

# converting the columns in to date time
cols = ['transactionDateTime','accountOpenDate','dateOfLastAddressChange']
df_transactions[cols] = df_transactions[cols].apply(pd.to_datetime, errors='coerce')
```

```python
# checking for number of unique values
for i in df_transactions.columns:
    print ('the number of unique values in   %s       is %d   and  dtype = %s' %(i,(len(df_transactions[i].unique())),(df_transactions[i].dtype)))
```

```python
# adding a column to find the difference in times of address change and account open date
df_transactions['diff_days'] = df_transactions['dateOfLastAddressChange'] - df_transactions['accountOpenDate']
df_transactions['diff_days'] = df_transactions['diff_days']/np.timedelta64(1,'D')
```

```python
def hr_func(ts):
    return ts.hour

def day_func(ts):
    return ts.day_name()

df_transactions['transaction_hour'] = df_transactions['transactionDateTime'].apply(hr_func)
df_transactions['transaction_day'] = df_transactions['transactionDateTime'].apply(day_func)
```

```python
# split the data in to dependent and independent variables (X,y)
X = df_transactions_copy.drop('isFraud',axis=1)
y = df_transactions_copy.isFraud
```

```python

from collections import Counter
from imblearn.over_sampling import ADASYN 

print('Original dataset shape {}'.format(Counter(y)))

ada = ADASYN(random_state=42)
X_res, y_res = ada.fit_resample(X, y)
print('Resampled dataset shape {}'.format(Counter(y_res)))
y_res.value_counts().plot(kind = 'bar')

#obtained variables (X_res, y_res)
```

```python
y_res.head()
```

```python
#splitting the data also, we have used stratifies because we wanted the same amount of fraud cases in the training dataset.
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.3, random_state=42,stratify= y_res)
```

```python
#initiating the model and importing the necessary libraries
from sklearn.linear_model import LogisticRegression

# initiating the mocel
model_lr = LogisticRegression()

# fitting the data
model_lr.fit(X_train,y_train)

#predecting on the test data
predictions_lr = model_lr.predict(X_test)
```

```python
# printing the performance metrics
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,roc_auc_score,roc_curve,recall_score
print('Accuracy score : %f' %(accuracy_score(y_test,predictions_lr)))
print('Precision score : %f' %((precision_score(y_test,predictions_lr))))
print(confusion_matrix(y_test,predictions_lr))


# plotting the AUC and measuring the ROC
fpr, tpr, _ = roc_curve(y_test, predictions_lr)
auc = roc_auc_score(y_test, predictions_lr)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
p = rf.predict(X_test)
```

```python
# printing the performance metrics
print('Accuracy score : %f' %(accuracy_score(y_test,p)))
print('Precision score : %f' %((precision_score(y_test,p))))
print(confusion_matrix(y_test,p))


# plotting the AUC and measuring the ROC
fpr, tpr, _ = roc_curve(y_test, p)
auc = roc_auc_score(y_test, p)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
